```python
import pandas as pd

# Laden des Datensatzes
df = pd.read_csv('python_challenge.csv')
df
```

## Task 1: Entfernen von Duplikaten

```python
# Anzahl der Zeilen vor dem Entfernen von Duplikaten speichern
original_row_count = len(df)
df = df.drop_duplicates()
# Index zurücksetzen
df = df.reset_index(drop=True)

# Anzahl der Zeilen nach dem Entfernen von Duplikaten speichern
deduplicated_row_count = len(df)

# Ergebnisse ausgeben
print(f"Anzahl der ursprünglichen Zeilen: {original_row_count}")
print(f"Anzahl der Zeilen nach Entfernen von Duplikaten: {deduplicated_row_count}")
# Überprüfen auf Duplikate
duplikate = df.duplicated().sum()

# Ausgabe der Anzahl von gefundenen Duplikaten
print(f"Überprüfen von Anzahl des Duplikaten im df: {duplikate}")
df
```

## Task 2

```python
import matplotlib.pyplot as plt

# Umwandlung des 'date' in das datetime-Format
df['date'] = pd.to_datetime(df['date'])

# Grafische Darstellung des Umsatzes über die Zeit für jede Filiale
for location in df['filiale'].unique():
    subset = df[df['filiale'] == location]
    subset.groupby('date')['umsatz'].sum().plot(label=location)

plt.legend()
plt.title('Umsatzentwicklung über die Zeit nach Filiale')
plt.xlabel('Datum')
plt.ylabel('Umsatz')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Task 3 

```python
# Berechnung des durchschnittlichen Tagesgewinns pro Produkt
avg_profit = df.groupby('product_name')['gewinn'].mean()

# Darstellung
avg_profit.plot(kind='bar', color='orange')
plt.title('Durchschnittlicher Tagesgewinn pro Produkt')
plt.xlabel('Produkt')
plt.ylabel('Durchschnittsgewinn')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Task 4

```python
df['date'] = pd.to_datetime(df['date'])

# Berechnung des Gesamtumsatzes pro Tag
total_revenue_per_day = df.groupby('date')['umsatz'].sum()

# Überprüfung der Datenpunkte
print(total_revenue_per_day)

# Plot der täglichen Umsatzdaten
total_revenue_per_day.plot(color='red', marker='o', linestyle='-', linewidth=2)
plt.title('Täglicher Gesamtumsatz')
plt.xlabel('Datum')
plt.ylabel('Umsatz')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Berechnung des gleitenden Durchschnitts mit einem 3-Tage-Fenster
moving_average = total_revenue_per_day.rolling(window=3, min_periods=1).mean()

# Grafische Darstellung des gleitenden Durchschnitts
moving_average.plot(color='green', marker='o', linestyle='-', linewidth=2)
plt.title('3-Tage Gleitender Durchschnitt des Gesamtumsatzes')
plt.xlabel('Datum')
plt.ylabel('Durchschnittlicher Umsatz')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```
